```python
#25 biggest countries in world

df1 = countries[['Country (or dependency)', 'Population (2020)', 'Density (P/Km²)']]
df2 = df1.nlargest(25, ['Population (2020)']) 
df2

#India 1382 million, Brazil 213 million, Turkey 84 million, UK 68 million, Italy 60 million
#South Korea, Romania and NL not in 25 biggest countries
```

```python
#plot of 25 biggest countries in the world

fig, ax = plt.subplots(figsize=(16, 6))
plt.xticks(rotation = 45)
sns.barplot(data = df2, x = 'Country (or dependency)', y = 'Population (2020)')
```
