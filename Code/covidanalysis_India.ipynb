```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import plotly.express as px
#import theme

%matplotlib inline
```

```python
COVID_CONFIRMED_URL = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'

covid_confirmed = pd.read_csv(COVID_CONFIRMED_URL)

print(covid_confirmed.shape)

covid_confirmed.head()
```

```python
covid_confirmed.columns
```

```python
COVID_DEATHS_URL = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'

covid_deaths = pd.read_csv(COVID_DEATHS_URL)

print(covid_confirmed.shape)

covid_deaths.head()
```

```python
COVID_RECOVERED_URL = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv'

covid_recovered = pd.read_csv(COVID_RECOVERED_URL)

print(covid_recovered.shape)

covid_recovered.head()
```

```python
covid_confirmed[['Province/State']] = covid_confirmed[['Province/State']].fillna('')
covid_confirmed.fillna(0, inplace=True)

covid_deaths[['Province/State']] = covid_deaths[['Province/State']].fillna('')
covid_deaths.fillna(0, inplace=True)

covid_recovered[['Province/State']] = covid_recovered[['Province/State']].fillna('')
covid_recovered.fillna(0, inplace=True)
```

```python
#check if NaN gone
covid_confirmed.isna().sum().sum()
```

```python
#NL confirmed
covid_confirmed_India = covid_confirmed[(covid_confirmed['Country/Region'] == 'India')]
covid_confirmed_India
#& 
```

```python
#NL deaths
covid_deaths_India = covid_deaths[(covid_deaths['Country/Region'] == 'India')]
covid_deaths_India
```

```python
#NL recovered
covid_recovered_India = covid_recovered[(covid_recovered['Country/Region'] == 'India')]
covid_recovered_India
```

```python
#nr confiremd cases India
covid_confirmed_India_count = covid_confirmed_India.iloc[:, 4:].sum().max()
covid_confirmed_India_count
```

```python
covid_deaths_India_count = covid_deaths_India.iloc[:, 4:].sum().max()
covid_deaths_India_count
```

```python
# percentage died compared to cases
covid_deaths_India_count/covid_confirmed_India_count
```

```python
#xx
covid_recovered_India_count = covid_recovered_India.iloc[:, 4:].sum().max()
covid_recovered_India_count
```

```python
covid_India_confirmed = covid_confirmed_India.iloc[:, 4:].sum(axis=0)

covid_India_confirmed.tail()
```

```python
covid_India_deaths = covid_deaths_India.iloc[:, 4:].sum(axis=0)

covid_India_deaths.tail()
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

sns.lineplot(x=covid_India_confirmed.index, y=covid_India_confirmed, sort=False, linewidth=2)
sns.lineplot(x=covid_India_deaths.index, y=covid_India_deaths, sort=False, linewidth=2)

ax.lines[0].set_linestyle("--")

plt.suptitle("COVID-19 India cases over the time evolution", fontsize=16, fontweight='bold', color='white')

plt.xticks(rotation=45)
plt.ylabel('Number of cases')

ax.legend(['Confirmed', 'Deaths'])

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 6))
ax.set(yscale="log")
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))

sns.lineplot(x=covid_India_confirmed.index, y=covid_India_confirmed, sort=False, linewidth=2)
sns.lineplot(x=covid_India_deaths.index, y=covid_India_deaths, sort=False, linewidth=2)

ax.lines[0].set_linestyle("--")

plt.suptitle("COVID-19 India cases over the time", fontsize=16, fontweight='bold', color='white')
plt.title("(logarithmic scale)", color='white')

plt.xticks(rotation=45)
plt.ylabel('Number of cases')

ax.legend(['Confirmed', 'Deaths'])

plt.show()
```
