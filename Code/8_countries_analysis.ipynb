```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

%matplotlib inline
```

```python
countries = pd.read_csv('../data/population_by_country_2020.csv')
```

```python
COVID_CONFIRMED_URL = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'

covid_confirmed = pd.read_csv(COVID_CONFIRMED_URL)
```

```python
COVID_DEATHS_URL = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'

covid_deaths = pd.read_csv(COVID_DEATHS_URL)
```

```python
covid_confirmed_8 = covid_confirmed[covid_confirmed['Country/Region'].isin(['Brazil', 'India', 'Italy', 'Netherlands', 'Romania', 'Korea, South', 'Turkey', 'United Kingdom'])]
covid_confirmed_8
```

```python
covid_confirmed_8 = covid_confirmed_8.drop(covid_confirmed_8.index[4:8])
covid_confirmed_8
```

```python
covid_confirmed_8 = covid_confirmed_8.drop(covid_confirmed_8.index[7:17])
covid_confirmed_8
```

```python
covid_deaths_8 = covid_deaths[covid_deaths['Country/Region'].isin(['Brazil', 'India', 'Italy', 'Netherlands', 'Romania', 'Korea, South', 'Turkey', 'United Kingdom'])]
covid_deaths_8
```

```python
covid_deaths_8 = covid_deaths_8.drop(covid_deaths_8.index[4:8])
covid_deaths_8
```

```python
covid_deaths_8 = covid_deaths_8.drop(covid_deaths_8.index[7:17])
covid_deaths_8
```

```python
covid_confirmed_8_countries = covid_confirmed_8[['Country/Region', '10/14/20']]
covid_confirmed_8_countries
```

```python
#replace name of Korea, South

covid_confirmed_8_countries['Country/Region'] = covid_confirmed_8_countries['Country/Region'].replace(['Korea, South'], 'South Korea')
covid_confirmed_8_countries
#example code df['column name'] = df['column name'].replace(['old value'],'new value')
```

```python
#create dataframe for covid cases on 14-10-2020

covid_8_countries = covid_confirmed_8_countries.rename(columns={'10/14/20': 'Covid_Cases'})
covid_8_countries

#examplecode hoteldata = hoteldata.rename(columns={'WoonlandVanGasten':'Country'
```

```python
covid_deaths_8 = covid_deaths_8[['Country/Region', '10/14/20']]
covid_deaths_8
```

```python
#add deaths to covid 8 countries

covid_8_countries['Covid_Deaths'] = pd.Series(covid_deaths_8['10/14/20'])
covid_8_countries

#example code df['columnF'] = pd.Series(df1['columnF'])
```

```python
covid_8_countries['%_Deaths_per_Cases'] = covid_8_countries['Covid_Deaths']/covid_8_countries['Covid_Cases'] * 100
covid_8_countries
```

```python
countries_select = countries[['Country (or dependency)', 'Population (2020)', 'Density (P/Km²)', 'Med. Age', 'Urban Pop %']]
countries_charact = countries_select[countries_select['Country (or dependency)'].isin(['Brazil', 'India', 'Italy', 'Netherlands', 'Romania', 'South Korea', 'Turkey', 'United Kingdom'])]
countries_charact
```

```python
columns = ['Country/Region', 'Population_2020', 'Density_Pop/Km2', 'Med_Age', 'UrbanPop_%']
countries_charact.columns = columns
```

```python
covid_countries = pd.merge(covid_8_countries,countries_charact,on='Country/Region')
```

```python
covid_countries
```

```python
covid_countries['Cases_per_million'] = round(covid_countries['Covid_Cases'] / covid_countries['Population_2020'] * 1000000,0)
covid_countries['Deaths_per_million'] = round(covid_countries['Covid_Deaths'] / covid_countries['Population_2020'] * 1000000,0)
covid_countries
```

```python
covid_countries_sorted1 = covid_countries.sort_values('Cases_per_million', ascending = False)
covid_countries_sorted1
```

```python
#figure of covid cases per million per country

fig, ax = plt.subplots(figsize=(12, 4))
plt.xticks(rotation = 0)
ax.set_title('Covid cases per million for 8 countries')
sns.barplot(data = covid_countries_sorted1, x = 'Country/Region', y = 'Cases_per_million')
```

```python
covid_countries_sorted2 = covid_countries.sort_values('Deaths_per_million', ascending = False)
covid_countries_sorted2
```

```python
#figure of covid deaths per million per country

fig, ax = plt.subplots(figsize=(12, 4))
plt.xticks(rotation = 0)
ax.set_title('Covid deaths per million for 8 countries')
sns.barplot(data = covid_countries_sorted2, x = 'Country/Region', y = 'Deaths_per_million')
```
